```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate example data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=2.5, random_state=0, return_centers=False) # type: ignore

# Apply KMeans
kmeans = KMeans(n_clusters=4) 
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the result
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.6)
plt.title("K-Means Clustering")
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Data points
X = np.array([[2, 10], [2, 5], [8, 4], [5, 8], [7, 5], [6, 4], [1, 2], [4, 9]])

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=100, cmap='viridis', edgecolor='k', label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, marker='X', label='Centroids')

# Optional: Annotate each point
for i, (x, y) in enumerate(X):
    plt.text(x + 0.1, y, f"P{i+1}", fontsize=9)

plt.title("K-Means Clustering Visualization", fontsize=14)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
